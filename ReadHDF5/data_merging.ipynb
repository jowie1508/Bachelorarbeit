```python
import pandas as pd
import pickle 
```

```python
with open('Data/heatpump/data_heatpump_cleaned_v1.pkl', 'rb') as f:
    load_dict = pickle.load(f)

with open('Data/weather/data_weather_v1.pkl', 'rb') as f:
    weather_data = pickle.load(f)
```

```python
load_dict_sorted = {}
for key in  sorted(load_dict.keys(), key=lambda x: int(x.replace("SFH", ""))):
    load_dict_sorted[key] = load_dict[key].rename(columns={'P_TOT': 'P_{}'.format(key), 'PF_TOT': 'PF_{}'.format(key)})
    
df_load = pd.concat([load_dict_sorted[df] for df in load_dict_sorted], axis=1)
df_load
```

```python
weather_data = weather_data[weather_data.index>=df_load.index[0]]
```

```python
df = pd.concat([df_load, weather_data], axis=1)
df['minute'] = pd.to_datetime(df.index, unit='s').minute
df['hour'] = pd.to_datetime(df.index, unit='s').hour
df['day'] = pd.to_datetime(df.index, unit='s').day
df['month'] = pd.to_datetime(df.index, unit='s').month
df['year'] = pd.to_datetime(df.index, unit='s').year
df
```
